```python
a = W_att @ np.concatenate([(X @ W.T)[connections[0]], (X @ W.T)[connections[1]]], axis=1).T
a
```

Leaky ReLUの定義

```python
def leaky_relu(x, alpha=0.2):
    return np.maximum(alpha*x, x)

e = leaky_relu(a)
e
```

隣接行列を注意スコアに変換

```python
E = np.zeros(A.shape)
E[connections[0], connections[1]] = e[0]
E
```

W_alphaは正規化されたエッジの重み行列

```python
def softmax2D(x, axis):
    e = np.exp(x - np.expand_dims(np.max(x, axis=axis), axis))
    sum = np.expand_dims(np.sum(e, axis=axis), axis)
    return e / sum

W_alpha = softmax2D(E, 1)
W_alpha
```

Graph Attention Networkの定義．gat1はマルチヘッドアテンション．gat2はシングルヘッドアテンション．

```python
import torch
torch.manual_seed(1)
import torch.nn.functional as F
from torch_geometric.nn import GATv2Conv, GCNConv
from torch.nn import Linear, Dropout


def accuracy(y_pred, y_true):
    """Calculate accuracy."""
    return torch.sum(y_pred == y_true) / len(y_true)


class GAT(torch.nn.Module):
    def __init__(self, dim_in, dim_h, dim_out, heads=8):
        super().__init__()
        self.gat1 = GATv2Conv(dim_in, dim_h, heads=heads)
        self.gat2 = GATv2Conv(dim_h*heads, dim_out, heads=1)

    def forward(self, x, edge_index):
        h = F.dropout(x, p=0.6, training=self.training)
        h = self.gat1(h, edge_index)
        h = F.elu(h)
        h = F.dropout(h, p=0.6, training=self.training)
        h = self.gat2(h, edge_index)
        return F.log_softmax(h, dim=1)

    def fit(self, data, epochs):
        criterion = torch.nn.CrossEntropyLoss()
        optimizer = torch.optim.Adam(self.parameters(), lr=0.01, weight_decay=0.01)

        self.train()
        for epoch in range(epochs+1):
            optimizer.zero_grad()
            out = self(data.x, data.edge_index)
            loss = criterion(out[data.train_mask], data.y[data.train_mask])
            acc = accuracy(out[data.train_mask].argmax(dim=1), data.y[data.train_mask])
            loss.backward()
            optimizer.step()

            if(epoch % 20 == 0):
                val_loss = criterion(out[data.val_mask], data.y[data.val_mask])
                val_acc = accuracy(out[data.val_mask].argmax(dim=1), data.y[data.val_mask])
                print(f'Epoch {epoch:>3} | Train Loss: {loss:.3f} | Train Acc: {acc*100:>5.2f}% | Val Loss: {val_loss:.2f} | Val Acc: {val_acc*100:.2f}%')

    @torch.no_grad()
    def test(self, data):
        self.eval()
        out = self(data.x, data.edge_index)
        acc = accuracy(out.argmax(dim=1)[data.test_mask], data.y[data.test_mask])
        return acc

# Create the Vanilla GNN model
gat = GAT(dataset.num_features, 32, dataset.num_classes)
print(gat)

# Train
gat.fit(data, epochs=100)

# Test
acc = gat.test(data)
print(f'GAT test accuracy: {acc*100:.2f}%')
```

# Graph Attention Networks for tensorflow user

```python
import tensorflow as tf
import tensorflow_gnn as tfgnn
import networkx as nx
from torch_geometric.datasets import Planetoid
from torch_geometric.utils import to_networkx

# PyTorch Geometricでデータセットをロード
dataset = Planetoid(root=".", name="CiteSeer")
data = dataset[0]

# NetworkX形式に変換
graph_nx = to_networkx(data)

# TensorFlow GNN形式に変換
def nx_to_tfgnn(graph, features, labels):
    # ノード特徴の辞書を作成
    node_features = {str(i): features[i].tolist() for i in range(len(features))}
    node_labels = {str(i): labels[i].tolist() for i in range(len(labels))}

    # エッジリストを取得
    edge_list = [(str(u), str(v)) for u, v in graph.edges]

    # TensorFlow GNNグラフ形式を構築
    graph_schema = tfgnn.GraphSchema()
    builder = tfgnn.GraphTensorBuilder(graph_schema)
    builder.add_nodes(node_set_name="nodes", features={"features": tf.convert_to_tensor(list(node_features.values()))})
    builder.add_edges(edge_set_name="edges", source="nodes", target="nodes", adjacency=tfgnn.Adjacency.from_edges(edge_list))
    return builder.finalize()

tf_graph = nx_to_tfgnn(graph_nx, data.x.numpy(), data.y.numpy())
```

```python
from tensorflow.keras import layers

class GATModel(tf.keras.Model):
    def __init__(self, input_dim, hidden_dim, output_dim, num_heads=8, dropout_rate=0.6):
        super().__init__()
        self.gat1 = tfgnn.keras.layers.GATv2Conv(input_dim, hidden_dim, num_heads)
        self.dropout1 = layers.Dropout(dropout_rate)
        self.gat2 = tfgnn.keras.layers.GATv2Conv(hidden_dim * num_heads, output_dim, 1)
        self.dropout2 = layers.Dropout(dropout_rate)

    def call(self, inputs, training=False):
        x, edge_index = inputs
        x = self.dropout1(x, training=training)
        x = self.gat1((x, edge_index), training=training)
        x = tf.nn.elu(x)
        x = self.dropout2(x, training=training)
        x = self.gat2((x, edge_index), training=training)
        return tf.nn.log_softmax(x, axis=1)
```

```python
def get_masked_data(graph_tensor, mask):
    return tf.boolean_mask(graph_tensor['nodes']['features'], mask)
```

```python
model = GATModel(input_dim=data.x.shape[1], hidden_dim=16, output_dim=dataset.num_classes)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

# 訓練データを用意
train_mask = data.train_mask.numpy()
x_train = get_masked_data(tf_graph, train_mask)
y_train = tf.boolean_mask(data.y.numpy(), train_mask)

# モデル訓練
model.fit(x_train, y_train, epochs=100, batch_size=32)
```

```python
# テストデータ
test_mask = data.test_mask.numpy()
x_test = get_masked_data(tf_graph, test_mask)
y_test = tf.boolean_mask(data.y.numpy(), test_mask)

# モデル評価
test_accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy[1] * 100:.2f}%")
```
